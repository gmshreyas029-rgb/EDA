```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'Name': ['Shreyas', 'Anu', 'Rahul', 'Unknown', 'Kiran'],
    'Age': [20, 21, 20.5, 19, 22],
    'Salary': [25000, 28000, 30000, 28000, 40000],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'General']
}

df = pd.DataFrame(data)
```

```python
dept_salary = df.groupby('Department')['Salary'].mean()

plt.figure()
dept_salary.plot(kind='bar')
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure()
plt.plot(df['Age'], df['Salary'], marker='o')
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
dept_count = df['Department'].value_counts()

plt.figure()
plt.pie(dept_count, labels=dept_count.index, autopct='%1.1f%%')
plt.title("Department Distribution")
plt.show()
```

```python
plt.figure()
plt.hist(df['Age'], bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Age'], df['Salary'])
plt.title("Scatter Plot: Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
import pandas as pd

data = {
    'Name': ['Shreyas', 'Anu', 'Rahul', None, 'Kiran'],
    'Age': [20, 21, None, 19, 22],
    'Salary': [25000, None, 30000, 28000, '40000'],
    'Department': ['IT', 'HR', 'IT', 'Finance', None]
}

df = pd.DataFrame(data)

print("Before Cleaning:\n")
print(df)
```

```python
print(df.isnull().sum())

df['Name'] = df['Name'].fillna('Unknown')

df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

df['Salary'] = df['Salary'].fillna(df['Salary'].median())

df['Department'] = df['Department'].fillna('General')

df = df.drop_duplicates()
```

```python
print("\nAfter Cleaning:\n")
print(df)
```
